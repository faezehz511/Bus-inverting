```python
import numpy as np
```

```python
def binar(x):
    a=[]
    for i in x:
        a.append('{:064b}'.format(int(i)))
    return a
```

##### lognormal distribution

```python
normal = np.random.lognormal(mean=40.,sigma=.10,size=1000)
u2 = binar(normal.tolist())
u2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(normal, hist=False)

plt.show()
```

```python
array2=np.zeros((1000,64))
for i in range(1000):
    for j in range(64):
        array2[i][j]=int(u2[i][j])
```

```python
#calculate number of transition of normal distribution
t = 0
for i in range(0,999):
    for j in range(0,64):
        if(array2[i][j] != array2[i+1][j]):
            t = t+1
print('sum of transitions(lognormal) =',t)
```

```python
# calculate number of transition in bus inverting for normal distribution
transition=0

for i in range(0,999):
    for j in range(0,64):
        if(array2[i][j] != array2[i+1][j]):
            transition = transition + 1
    #print('Step transition [', i,'] is= ', transition)
    if(transition > 32):
        #print('hi') 
        for jp in range(0,64):
            if array2[i+1][jp]==0:
                array2[i+1][jp] = 1
            else:
                array2[i+1][jp] = 0
        #print(array1[i+1])
    transition = 0
t = 0
for i in range(0,999):
    for j in range(0,64):
        if(array2[i][j] != array2[i+1][j]):
            t = t+1
print('sum of transitions with bus-inverting(lognormal) =',t)
```
